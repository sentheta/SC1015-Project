```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
import warnings
from IPython.display import display
sb.set() 
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Housing.csv")
data.head()
```

```python
# Removing empty spaces from the columns
data.columns = data.columns.str.strip()

# Removing leading and trailing whitespaces
data = data.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Removing duplicate rows
data.drop_duplicates(inplace=True)

# Data types of columns
print("\nData types of columns:")
print(data.dtypes)

# Missing values
print("\nMissing values:")
print(data.isnull().sum())

# Dropping the rows with resale_price and floor_area_sqm as 0
data['resale_price'].fillna(0, inplace=True)
data = data.dropna(subset=['floor_area_sqm'])
data = data.dropna(subset=['resale_price'])
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv("Housing.csv")

# Convert the 'month' column to datetime format
df['month'] = pd.to_datetime(df['month'], format='%d-%m-%Y')

# Extract the year from the 'month' column and create a new column named 'year'
df['year'] = df['month'].dt.year

# Save the DataFrame back to a CSV file
df.to_csv('cleaned-Housing.csv', index=False)  # Set index=False to avoid saving the DataFrame index as a column
```
